```python
# Import libraries
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Open the JSON file and load the data
with open("../dataset/experiments.json") as f:
    data = json.load(f)

df = pd.DataFrame(data)

df.head()
```

```python
df.describe()
```

```python
# Scatter plot of latency vs error rate
sns.scatterplot(
    data=df,
    x="latencyMs",
    y="errorRate"
)

plt.title("Latency vs Error Rate")
plt.show()
```

```python
# Correlation matrix
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```
